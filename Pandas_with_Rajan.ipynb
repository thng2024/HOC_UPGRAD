üçé you are Asked to find all Players from 'Texas' College.

# üíª Handling Multiple Conditions
---

```python
nba.head(3)
```

üèÜ You are Aksed to Find Players from 'Texas' College Playing from 'PG' Position.

```python
c1 = nba.College  == 'Texas'
c2 = nba.Position == 'PG'

nba[ c1 & c2 ]             # Here & is Logical AND
```

```python
# here is how you do it using Direct Expression

nba[  (nba.College == 'Texas') & (nba.Position == 'PG' ) ]
```
